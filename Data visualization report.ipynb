```python

df = pd.DataFrame(df)

# Remove "Serial Number" and "|" along with numbers from the third column
df['Application Area'] = df['Serial Number | Application Area'].str.replace(r'\b\d+\s*\|\s*', '').str.strip()

# Drop the original column
df = df.drop(columns=['Serial Number | Application Area'])

# Print the result
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Split the methods and create a new row for each method
methods_df = df['MCDM Tools / Methods'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).to_frame('Method')
df_split = df.drop('MCDM Tools / Methods', axis=1).join(methods_df)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Application Area', data=df_split, hue='Method', palette='viridis')
plt.title('Distribution of MCDM Methods Across Application Areas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(df)

# Split the methods and create a new row for each method
methods_df = df['MCDM Tools / Methods'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).to_frame('Method')
df_split = df.drop('MCDM Tools / Methods', axis=1).join(methods_df)

# Count the occurrences of each method in each area
method_counts = df_split.groupby(['Application Area', 'Method']).size().reset_index(name='Count')

# Pivot the table for better visualization
method_counts_pivot = method_counts.pivot(index='Application Area', columns='Method', values='Count').fillna(0)
```

```python
# Create a heatmap to visualize the results
plt.figure(figsize=(12, 8))
sns.heatmap(method_counts_pivot, cmap='viridis', annot=True, fmt='g', cbar_kws={'label': 'Count'})
plt.title('Method Usage Across Application Areas')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you already have a DataFrame named df with columns 'MCDM Tools / Methods' and 'Serial Number | Application Area'

# Split the methods and create a new row for each method
methods_df = df['MCDM Tools / Methods'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).to_frame('Method')
df_split = df.drop('MCDM Tools / Methods', axis=1).join(methods_df)

# Count the occurrences of each method
method_counts = df_split['Method'].value_counts().reset_index(name='Count')

# Create a bar plot to visualize the results
plt.figure(figsize=(20, 20))
sns.barplot(x='Count', y='index', data=method_counts, palette='viridis')
plt.title('Distribution of MCDM Methods Across All Application Areas')
plt.xlabel('Count')
plt.ylabel('Method')
plt.tight_layout()
plt.show()
```
